We wanted to define a function that would allow to change easily the parameters of the model we wished to study. Therefore, in the next function (this is the one we will be using to select the best models), we have the opportunity to specify whether or not we want to rescale the data, leverage PCA, use Polynomial Features. It is important to specify that the choices we are going to make will be based on cross validation performances on the specific swing state (and not validation performance, which might be subject to 'lucky' splits).

```python
def modelling(X_train, y_train, models, scaler=True, Poly=1, PCA_comp=0):
    # pca is the total variance explained required. 0 means that we do not want to perform pca
    if 'State' in X_train.columns:
        X_train = X_train.drop('State', axis=1)
    if 'Year' in X_train.columns:
        X_train = X_train.drop('Year', axis=1)
    if PCA_comp > 0 and not scaler:
        raise ValueError('We must normalize before performing PCA')
    if scaler:
        # print('Normalizing the data ...')
        scaler = MinMaxScaler().fit(X_train)
        X_train = scaler.transform(X_train)
    if PCA_comp > 0:
        # print('Performing PCA ...')
        pca = PCA(n_components=X_train.shape[1])
        pca.fit(X_train)
        X_train = pca.transform(X_train)
        total_variance = np.cumsum(pca.explained_variance_ratio_)
        for i, variance in enumerate(total_variance):
            if variance > PCA_comp:
                break 
        X_train = X_train[:, :i+1]
    if Poly > 1:
        # print('Polynomial transformation ...')
        poly_features = PolynomialFeatures(degree=Poly, include_bias=False).fit(X_train)
        X_train = poly_features.transform(X_train)
    model_cv_accuracies = []
    # print('Modelling and gathering the predictions ..')
    for i, model in enumerate(models):
        model_cv = cross_validate(model, X_train, y_train, cv=5, scoring='accuracy')
        model_cv_accuracy = np.mean(model_cv['test_score'])
        model_cv_accuracies.append(model_cv_accuracy)
    return models, model_cv_accuracies
```

We have done extensive model selection in the sections above. Nevetheless, we have identified a flaw with the above modeling approach. The flaw arises from the construction of the cross-validation set. Since for every year the data set contains two observations (one for the democrat candidate, one for the republican) if once of the two observations is in the training and the oter in the CV sets respectively than the model will easily recognize and memorize the pattern and therefore the model becomes near deterministic, i.e. there is a large information leakage. 

In order to prevent this issue, the first approach we will try is to get rid of half the observations, in other words we will only consider the observations for one (the republican candidate) for each election year.

For one swing-state, we need to take into account the correlation with every other state and insert the data from the other states with the correct weight. They will be contained into final_importance_df (dictionnary with keys being the swing states).

```python
final_importance_dfs = {}
for index_state in index_swing_states:  # model only for the swing states
    data_swing_state = []
    for i, state in enumerate(all_states):  
        if state == all_states[index_state]:
            continue
        if state != all_states[index_state]:   # for all the state but the swing state, remove the 2020 election
            data_state = final_df[final_df['State'] == state]
            data_state = data_state.iloc[:-1, :]
        correlation = int(correlation_matrix.iloc[index_state, i])
        if correlation > 0:  # add correlation * the dataframe for the state (where we removed the 2020 election)
            data_state_importance = pd.concat([data_state for _ in range(correlation)])
            if data_state.values.tolist():
                try:
                    data_swing_state = pd.concat([data_swing_state, data_state_importance])
                except:
                    data_swing_state = data_state_importance
    final_importance_dfs[all_states[index_state]] = data_swing_state
# for every swing state, the corresponding values are missing, this is in order to help for the validation
```

We found an efficient way to perform cross-validation (and not simply validation), we will manually create the folds over the validation state (ie the specific swing state), resulting in differents training and validation sets (3 to be specific). Then, we will train the same model on every different split and aggregate the different results in order to get the cross-validation performance.

```python
swing_state = 'Texas' # example
```

```python
def split(initial_df, final_importance_df, swing_state): # particular state missing
    swing_state_data = initial_df[initial_df['State'] == swing_state]  # data to be protected
    data_test = swing_state_data[swing_state_data['Year'] == 2020].iloc[:1, :] # test data
    X_test, y_test = data_test.drop('Result', axis=1), data_test['Result'] # X test and test labels  
    swing_state_data = swing_state_data[swing_state_data['Year'] != 2020] # we remove the test data
    kf = KFold(n_splits=3, random_state=1, shuffle=True)
    Xs, X_vals, ys, yvals = [], [], [], []
    for train_index, val_index in kf.split(swing_state_data):
        X_train, X_val = swing_state_data.iloc[train_index, :], swing_state_data.iloc[val_index, :]
        X_val, y_val = X_val.drop('Result', axis=1), X_val['Result']
        X_vals.append(X_val)
        yvals.append(y_val)
        df_corr_to_concat = pd.concat([X_train for _ in range(10)])
        X_values = pd.concat((final_importance_df[final_importance_df['State'] != swing_state], df_corr_to_concat))
        X_values = X_values.sample(frac=1, replace=False) 
        X, y = X_values.drop('Result', axis=1), X_values['Result']
        Xs.append(X)
        ys.append(y)
    return Xs, X_vals, X_test, ys, yvals, y_test
```

We wanted to define a function that would allow to change easily the parameters of the model we wished to study. Therefore, in the next function (this is the one we will be using to select the best models), we have the opportunity to specify whether or not we want to rescale the data, leverage PCA, use Polynomial Features. It is important to specify that the choices we are going to make will be based on cross validation performances on the specific swing state (and not validation performance, which might be subject to 'lucky' splits).

```python
def modelling(X_train, X_val, y_train, y_val, models, scaler=True, Poly=1, PCA_comp=0):
    # pca is the total variance explained required. 0 means that we do not want to perform pca
    if 'State' in X_train.columns:
        X_train = X_train.drop('State', axis=1)
        X_val = X_val.drop('State', axis=1)
    if 'Year' in X_train.columns:
        X_train = X_train.drop('Year', axis=1)
        X_val = X_val.drop('Year', axis=1)
    if 'republican' in X_train.columns:
        X_train = X_train.drop('republican', axis=1)
        X_val = X_val.drop('republican', axis=1)
    if PCA_comp > 0 and not scaler:
        raise ValueError('We must normalize before performing PCA')
    # if PCA_comp > 1 and Poly > 1:
        # print('Performing Polynomial Transformation on top of PCA ...')
    if scaler:
        # print('Normalizing the data ...')
        scaler = MinMaxScaler().fit(X_train)
        X_train = scaler.transform(X_train)
        X_val = scaler.transform(X_val)
    if PCA_comp > 0:
        # print('Performing PCA ...')
        pca = PCA(n_components=X_train.shape[1])
        pca.fit(X_train)
        X_train = pca.transform(X_train)
        X_val = pca.transform(X_val)
        total_variance = np.cumsum(pca.explained_variance_ratio_)
        for i, variance in enumerate(total_variance):
            if variance > PCA_comp:
                break 
        X_train, X_val = X_train[:, :i+1], X_val[:, :i+1]
    if Poly > 1:
        # print('Polynomial transformation ...')
        poly_features = PolynomialFeatures(degree=Poly, include_bias=False).fit(X_train)
        X_train = poly_features.transform(X_train)
        X_val = poly_features.transform(X_val)
    accuracies = []
    # print('Modelling and gathering the predictions ..')
    for i, model in enumerate(models.copy()):
        model.fit(X_train, y_train)
        predictions = model.predict(X_val)
        accuracies.append(accuracy_score(y_val, predictions))
    return accuracies
```

## Predictions

This first function allow to create the train/test split for every state (we will use it only for the swing states). It is different from the previous ones because we included the same PCA degree and the sampe degree for polynomial features (after having explored which models perform best).

```python
def train_test_split_modelling(initial_df, final_importance_df, swing_state): # particular state missing
    swing_state_data = initial_df[initial_df['State'] == swing_state]  # data to be protected
    data_test = swing_state_data[swing_state_data['Year'] == 2020].iloc[:1, :] # test data
    X_test, y_test = data_test.drop('Result', axis=1), data_test['Result'] # X test and test labels  
    swing_state_data = swing_state_data[swing_state_data['Year'] != 2020] # we remove the test data
    if 'Republican' in swing_state_data.columns:
        swing_state_data.drop('Republican', axis=1, inplace=True)
    values_swing_state = final_importance_df[swing_state]
    if 'Republican' in values_swing_state.columns:
        values_swing_state.drop('Republican', axis=1, inplace=True)
    df_corr_to_concat = pd.concat([swing_state_data for _ in range(10)])
    X_values = pd.concat((final_importance_df[swing_state], df_corr_to_concat))
    X_values = X_values.sample(frac=1, replace=False) 
    X_train, y_train = X_values.drop('Result', axis=1), X_values['Result']
    if 'State' in X_train.columns:
        X_train = X_train.drop('State', axis=1)
        X_test = X_test.drop('State', axis=1)
    if 'Year' in X_train.columns:
        X_train = X_train.drop('Year', axis=1)
        X_test = X_test.drop('Year', axis=1)
    if 'republican' in X_train.columns:
        X_train = X_train.drop('republican', axis=1)
        X_test = X_test.drop('republican', axis=1)
    #scale data
    scaler = MinMaxScaler().fit(X_train)
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)
    #pca
    pca = PCA(n_components=X_train.shape[1])
    pca.fit(X_train)
    X_train = pca.transform(X_train)
    X_test = pca.transform(X_test)
    total_variance = np.cumsum(pca.explained_variance_ratio_)
    for j, variance in enumerate(total_variance):
            if variance > 0.8:
                break 
    X_train, X_test = X_train[:, :j+1], X_test[:, :j+1]
    #poly
    poly_features = PolynomialFeatures(degree=3, include_bias=False).fit(X_train)
    X_train = poly_features.transform(X_train)
    X_test = poly_features.transform(X_test)
    return X_train, X_test, y_train, y_test, pca
```

#### Additional EDA

One last thing we wanted to do with this model is to corroborate the fact that the swing states were 'swing-states' in the statistical term (and not just in common knowledge): we wanted to know if our model was certain of its predictions on these swing states.

```python
bootstraping_probabilities = {}
n_bootstraps = 100
for state in swing_states:
    X_train, X_test, y_train, y_test, pca = train_test_split_modelling(final_df, final_importance_dfs, state)
    pcas.append(pca)
    if state in swing_states_lr1:
        model = LogisticRegression(C=0.1, max_iter=10000, penalty='l1', solver='liblinear')
    if state in swing_states_lr2:
        model = LogisticRegression(C=0.001, max_iter=10000)
    if state in swing_states_knn:
        model = KNeighborsClassifier()
    bootstraping_state = []
    for _ in range(n_bootstraps):
        idx = list(np.random.choice(len(X_train), size = len(X_train), replace=True))
        X_boot = X_train[idx]
        y_boot = y_train.values[idx]
        model.fit(X_boot, y_boot)
        bootstraping_state.append(model.predict_proba(X_test)[0][1])
    bootstraping_probabilities[state] = bootstraping_state
```

```python
to_predict = final_df[final_df['Year'] != 2020]
to_predict.drop('republican', axis=1, inplace=True)
test = final_df[final_df['Year'] == 2020]
probabilities_model_2 = []
test_proba = []
for i, state in enumerate(all_states):
    if state in swing_states:
        index = swing_states.index(state)
        test_state = test[test['State'] == state]
        data_state = to_predict[to_predict['State'] == state]
        data_state.drop('State', axis=1, inplace=True)
        test_state.drop('State', axis=1, inplace=True)
        data_state.drop('Year', axis=1, inplace=True)
        test_state.drop('Year', axis=1, inplace=True)
        X, y = data_state.drop('Result', axis=1), data_state['Result']
        X_test, y_test = test_state.drop('Result', axis=1), test['Result']
        X_test.drop('republican', axis=1, inplace=True)
        scaler = MinMaxScaler().fit(X)
        X_train = scaler.transform(X)
        swing_state_data = final_df[final_df['State'] == state]  # data to be protected
        swing_state_data = swing_state_data[swing_state_data['Year'] != 2020] # we remove the test data
        if 'Republican' in swing_state_data.columns:
            swing_state_data.drop('Republican', axis=1, inplace=True)
        values_swing_state = final_importance_dfs[swing_state]
        if 'Republican' in values_swing_state.columns:
            values_swing_state.drop('Republican', axis=1, inplace=True)
        df_corr_to_concat = pd.concat([swing_state_data for _ in range(10)])
        X_values = pd.concat((final_importance_dfs[swing_state], df_corr_to_concat))
        X_values = X_values.sample(frac=1, replace=False) 
        X_train_fit, _ = X_values.drop('Result', axis=1), X_values['Result']
        if 'State' in X_train_fit.columns:
            X_train_fit = X_train_fit.drop('State', axis=1)
        if 'Year' in X_train_fit.columns:
            X_train_fit = X_train_fit.drop('Year', axis=1)
        if 'republican' in X_train_fit.columns:
            X_train_fit = X_train_fit.drop('republican', axis=1)
        #scale data
        scaler = MinMaxScaler().fit(X_train_fit)
        X_train_fit = scaler.transform(X_train_fit)
        X_test = scaler.transform(X_test)
        pca = PCA(n_components=X_train_fit.shape[1])
        pca.fit(X_train_fit)
        X_train = pca.transform(X_train)
        X_test = pca.transform(X_test)
        total_variance = np.cumsum(pca.explained_variance_ratio_)
        for j, variance in enumerate(total_variance):
                if variance > 0.8:
                    break 
        X_train = X_train[:, :j+1]
        X_test = X_test[:, :j+1]
        poly_features = PolynomialFeatures(degree=3, include_bias=False).fit(X_train)
        X_train = poly_features.transform(X_train)
        X_test = poly_features.transform(X_test)
        model = models[index]
        probability = model.predict_proba(X_train)[:, 1]
        test_probab = model.predict_proba(X_test)[:, 1]
        test_proba.append(test_probab)
    else:
        probability = 1*(to_predict[to_predict['State'] == state]['rep_loyalty'].values > 0.5)
        test_probab = 1*(test[test['State'] == state]['rep_loyalty'].values > 0.5)
        test_proba.append(test_probab.tolist())

    probabilities_model_2.append(probability.tolist())

probabilities_model_2 = np.array([ll for l in probabilities_model_2 for ll in l])
test_proba = np.array([ll for l in test_proba for ll in l])
    
    
```

```python
X_train_final = np.array([[proba_1, proba_2] for proba_1, proba_2 in zip(probabilities_model_1, probabilities_model_2)])
lr = LogisticRegression().fit(X=X_train_final, y=y_train_final)
final_test = np.array([[proba_1, proba_2] for proba_1, proba_2 in zip(y_pred_model1, test_proba)])
final_accuracy = lr.score(final_test, y_test)
print('The score of the final Ensemble Learning is ', final_accuracy)
final_predictions = lr.predict(final_test)
```
